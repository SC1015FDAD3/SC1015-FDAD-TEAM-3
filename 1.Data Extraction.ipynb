# Create Joint DataSet

```python
import pandas as pd

jointdataset = pd.DataFrame()

for i in range ( 2019 , 2023 ) :
    name = 'TikTok_songs_'
    filename = name + str(i) + '.csv'
    data = pd.read_csv(filename) ;
    year = []
    for j in range ( 0 , data.shape[0] ) :
        year += [i]
    data['Year'] = year 
    jointdataset = pd.concat([ data , jointdataset ] , axis = 0 )

jointdataset.to_csv("Full_DataSet.csv" , index = False )
```
